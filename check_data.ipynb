```python
import os
from glob import glob
```

```python
path = '/mnt/hdd/dataset/uav_dataset/val/20190925_101846_1_1/visible/'

im_list = sorted(glob(path + '*.jpg'))
mask_list = sorted(glob(path + '*.npy'))
print(len(im_list), len(mask_list), im_list[0:5], mask_list[0:5])
```

```python
for path in im_list:
    filename = os.path.splitext(path)[0]
    found = False

    for mask in mask_list:
        fn = os.path.splitext(mask)[0]
        # print(fn[-4:], filename[-9:-5])

        if fn[-9:-5] == filename[-4:]:
            found = True
    
    if found == False:
        print('Found:', filename)
        # print(fn[-9:-5])
    
    # print(filename[-5:])
```
